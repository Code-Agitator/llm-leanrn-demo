```python
import tiktoken
```

```python
tokenizer = tiktoken.encoding_for_model('gpt-3.5-turbo')
tokenizer.name
```

```python
res = tokenizer.encode('tiktoken is greet!')
print(len(res))
print(res)
```

```python
[tokenizer.decode_single_token_bytes(i) for i in res]
```

从输出中看出，token数量并不是单词的数量，部分单词被切分为多个token，比如'tiktoken'被切分为'tik'和'token‘多个数字
